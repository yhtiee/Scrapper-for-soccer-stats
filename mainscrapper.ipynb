```python
import requests
from bs4 import BeautifulSoup
url= "https://www.soccerstats.com/matches.asp"
response= requests.get(url)
print (response)
soup = BeautifulSoup(response.content,"html.parser")
```

```python
home = []
away = []
match = soup.find_all(class_="steam")
for x in match:
    matches = [x.text for x in match]
# print(matches)
for a in range(len(matches)):
    if a % 2 == 0 :
        home.append(matches[a])
    else:
        away.append(matches[a])
join = list(zip(home, away))
print(join)
```
